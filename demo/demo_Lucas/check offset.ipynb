```python
from PIL import Image
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
```

```python
f_img_o = Path("./DexiNed/BIPEDv2/BIPED/edges/imgs/test/rgbr/RGB_008.jpg")
f_img_gt = Path("./DexiNed/BIPEDv2/BIPED/edges/edge_maps/test/rgbr/RGB_008.png")
```

```python
img_o = Image.open(f_img_o).convert("RGB")
img_gt = Image.open(f_img_gt).convert("L")
# img_o = np.array(Image.open(f_img_o).convert("RGB"))
# img_gt = np.array(Image.open(f_img_gt).convert("L"))

# 将边界线转换为 RGBA（红色）
edges_colored = Image.new("RGBA", img_o.size, (255, 0, 0, 0))  # 创建透明图
edges_colored.putalpha(img_gt)  # 让边界线变成红色的透明层

# 叠加到原图
image = img_o.convert("RGBA")  # 确保原图也是 RGBA 格式
combined = Image.alpha_composite(image, edges_colored)  # 叠加
```

```python
plt.imshow(np.array(combined.convert("RGB")))
plt.show()
```
